```python
############ Important GRIDSEARCHCV ############
 


import numpy as np
from keras import models
from keras import layers
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_classification



def create_network(optimizer='rmsprop'):
    
    classifier = models.Sequential()
    classifier.add(Conv2D(20, (5, 5), input_shape = (64, 64, 1), activation = 'relu'))
    classifier.add(MaxPooling2D(pool_size = (2, 2)))
    classifier.add(Conv2D(40, (5, 5), input_shape = (64, 64, 1), activation = 'relu'))
    classifier.add(MaxPooling2D(pool_size = (2, 2)))
    classifier.add(Conv2D(80, (4, 4), input_shape = (64, 64, 1), activation = 'relu'))
    classifier.add(MaxPooling2D(pool_size = (2, 2)))
    classifier.add(Conv2D(80, (4, 4), input_shape = (64, 64, 1), activation = 'relu'))
    classifier.add(MaxPooling2D(pool_size = (2, 2)))
    classifier.add(Flatten())
    classifier.add(Dense(units = 80, activation = 'relu'))
    classifier.add(Dense(units = 64, activation = 'relu'))
    classifier.add(Dense(units = 1, activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    
    return classifier

neural_network = KerasClassifier(build_fn=create_network, verbose=0)

epochs = [2,5, 10]
batches = [5, 10,32, 100]
optimizers = ['rmsprop', 'adam']

# Create hyperparameter options
hyperparameters = dict(optimizer=optimizers, epochs=epochs, batch_size=batches)


# Create grid search
grid = GridSearchCV(estimator=neural_network, cv=3, param_grid=hyperparameters)

# Fit grid search
grid_result = grid.fit(xz, yy)


grid_result.best_params_



##### Important GRIDSEARCHCV
```

```python
classifier = models.Sequential()
classifier.add(Conv2D(20, (5, 5), input_shape = (64, 64, 1), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Conv2D(40, (5, 5), input_shape = (64, 64, 1), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Conv2D(80, (4, 4), input_shape = (64, 64, 1), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Conv2D(80, (4, 4), input_shape = (64, 64, 1), activation = 'relu'))
classifier.add(MaxPooling2D(pool_size = (2, 2)))
classifier.add(Dropout(0.3))
classifier.add(Flatten())
classifier.add(Dense(units = 80, activation = 'relu'))
classifier.add(Dense(units = 64, activation = 'relu'))
classifier.add(Dense(units = 1, activation = 'sigmoid'))
classifier.compile(optimizer = 'rmsprop', loss = 'binary_crossentropy', metrics = ['accuracy'])
batchsize = 5
num_epochs = 3
best_model = 'cnn_best.h5'
last_model = 'cnn_model.h5'
# #Training Model # for error will be min
checkpointer = ModelCheckpoint(filepath = best_model, monitor = 'val_acc', verbose = 0, save_best_only = True, mode = 'max')
classifier.fit(X, Y, batch_size = batchsize, epochs = num_epochs, verbose = 1, validation_split = 0.2, callbacks = [checkpointer])
#classifier.fit(X_train, y_train, batch_size = batchsize, epochs = num_epochs, verbose = 1, validation_split = 0.2, callbacks = [checkpointer])
```
